```python
filelist = []
for i in range(1,287):
    a = '%0.6d' % i
    filelist.append(a+'.txt')
```

```python
filelist
```

```python
for i in filelist:
    try:
        f = open(i,'r')
        c = f.readlines()
        l2 = c[1]
        f.close()
        f = open(i,'w')
        f.write('0\n')
        f.write(l2)
        f.close()
    except Exception as e:
        print (e)
        print (i)
```
